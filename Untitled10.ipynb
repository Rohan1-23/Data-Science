# Defuzzication of reliability part

```python
def Converting_into_7values(fuzzy_members):
    values=np.array(fuzzy_members)
    list=[]
    n=len(values)
    for i in range(n-1):
        a=(values[i]+values[i+1])/2
        list.append([values[i],a])
    list.append([values[n-1]])
     
    flat_list = []
    for sublist in list:
        for item in sublist:
            flat_list.append(item) 
     
         
    return flat_list

dcp=(0,0,1,2)     
print(Converting_into_7values(dcp))
```

```python
def defuzzy (reliability):
    array=np.array(reliability)
    mean=np.mean(array)
    return mean
print(defuzzy ((0,0.1,0,3)))
```
